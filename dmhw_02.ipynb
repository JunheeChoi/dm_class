```python
#02
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV 파일을 데이터프레임으로 읽어옴
housing_df = pd.read_csv('https://raw.githubusercontent.com/reisanar/datasets/master/BostonHousing.csv', usecols=['CRIM', 'RM', 'AGE', 'LSTAT', 'MEDV'])
housing_df.head()
```

```python
# 변수들 간의 상관관계 계산
correlations = housing_df.corr().round(2)

# 히트맵 생성
plt.figure(figsize=(8, 6))
sns.heatmap(data=correlations, annot=True,cmap = 'coolwarm')
plt.title('Correlation Heatmap of Boston Housing Data', fontsize=16)
plt.show()
```

## 해당 히트맵을 통해 각 변수들 간의 상관관계를 파악할 수 있음
### MEDV(주택 가격 중앙값)은 RM(방의 수)과 양의 상관관계를 가짐
### LSTAT(하위계층의 비율)은 MEDV와 음의 상관관계를 가짐
### CRIM(인구 1인당 범죄 발생 수)은 MEDV 와 강한 음의 상관관계를 가짐
