# Datamining 11주차 - 군집분석

## 군집 분석 : 개요
* 대표적인 비지도 학습 방법
* 인사이트를 얻기 위해 데이터를 동질적인 군집들로 세분화하는 것
* 데이터 셋을 동질적인 레코드로 구성된 군지븡로 분리 하는 것 -> 전체의 이질적인 데이터 셋 보다는 각 군집을 개별적으로 모델링 -> 지도 학습의 정능을 향상
* 계층적 군집화 & k-평균 군집화
* ** 계층적 군집화 : 레코드 간의 거리와 군집 간의 거리에 기초하여 순차적으로 가까운 레코드를 합침. 덴드로그램
* ** k-평균 군집화 : 레코드들은 각 군집 간 거리에 의해 미리 정해진 군집들 중 하나로 할당
* 관측치의 다양한 특성 변수의 거리를 이용해 유사한 그룹을 찾는 방법
* 탐색적이고 기술적인 분석 방법의 하나로 반응 변수가 없거나 반응 변수를 선정하기 전 단계에서 데이터 특성 파악 목적으로 주로 활용
* VS PCA :
* ** 군집 분석은 관측치 혹은 변수의 유사성을 기반으로 서로 비슷한 그룹으로 묵는 방법인 반면, 주성분 분석은 주어진 데이터의 특성을 잘 나타내는 저차원의 선형 결합을 찾는 방법임. 군집 분석은 주로 관측치들을 묶는 데 많이 이용되구 주성분 분석은 변수들을 묶는 데 주로 이용함
* 분할적 군집 분석 : 군닙의 수를 미리 결정한 뒤 군집화를 수행하는 방법임. 이때 각 군집이 서로 겹치지 않게 군집의 수를 결정하기 때문에 분할적 군집 분석이라고 함. 계층적 군집 방법에 비해 계산량이 적기 때문에 대용량 데이터 분석에 적합한 방법 (예 : k-means clustering)
* 계층적 군집 분석 : 한 군집 내 부분 군집을 허용하는 방법으로 각각의 관측치를 하나의 군집으로 간주하고 데이터의 특성이 가까운 부분 군집끼리 순차적으로 결합해 나가는 병합적 방법 (Agglomerative method) 이 주로 사용됨. 덴드로그램을 이용한 시각화를 통해 각 관측치가 어떤 관측치와 가까운지 파악할 수 있어 군집들을 자연적인 계층 구조로 분석하고자 할 때 유용함

### 군집 분석 : 주요 이슈
* 군집 분석에 활용할 변수들의 척도 차이를 확인. 군집 분석은 거리를 기반으로 군집을 형성하기 때문에 활용 변수의 척도 차이가 큰 경우 다ㄴ위가 큰 변수에 가중치가 부여되어 군집이 왜곡될 수 있음 -> 각 변수의 척도 차이가 존재하면 분석에 앞서 정규화 또는 표준화 작업을 수행해야 함
* 계층적 군집은 어떤 유사성 측도와 연결 방법으로 이용할 것인지 결정하는 것이 군집 형성에 큰 영향을 줄 수 있음. 또한 군집 결정을 위한 어디에서 덴드로 그램의 절단 계층을 선택하는 것도 중요한 의사결정 요인
* 분할적 군집 형성 방법 중 하나인 k=평균 군집 분석은 k 값을 어떻게 지정해줄 것인가에 대한 선택이 군집 형성에 매우 큰 영향을 줌. k- 평균 군집 분석 역시 각 관측치 간 유사성 측도를 무엇으로 할지 결정하는 것 또한 중요함
