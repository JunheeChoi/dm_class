## 군집 분석 : 개요
* 대표적인 비지도 학습 방법
* 인사이트를 얻기 위해 데이터를 동질적인 군집들로 세분화하는 것
* 데이터 셋을 동질적인 레코드로 구성된 군지븡로 분리 하는 것 -> 전체의 이질적인 데이터 셋 보다는 각 군집을 개별적으로 모델링 -> 지도 학습의 정능을 향상
* 계층적 군집화 & k-평균 군집화
* ** 계층적 군집화 : 레코드 간의 거리와 군집 간의 거리에 기초하여 순차적으로 가까운 레코드를 합침. 덴드로그램
* ** k-평균 군집화 : 레코드들은 각 군집 간 거리에 의해 미리 정해진 군집들 중 하나로 할당
* 관측치의 다양한 특성 변수의 거리를 이용해 유사한 그룹을 찾는 방법
* 탐색적이고 기술적인 분석 방법의 하나로 반응 변수가 없거나 반응 변수를 선정하기 전 단계에서 데이터 특성 파악 목적으로 주로 활용
* VS PCA :
* ** 군집 분석은 관측치 혹은 변수의 유사성을 기반으로 서로 비슷한 그룹으로 묵는 방법인 반면, 주성분 분석은 주어진 데이터의 특성을 잘 나타내는 저차원의 선형 결합을 찾는 방법임. 군집 분석은 주로 관측치들을 묶는 데 많이 이용되구 주성분 분석은 변수들을 묶는 데 주로 이용함
* 분할적 군집 분석 : 군닙의 수를 미리 결정한 뒤 군집화를 수행하는 방법임. 이때 각 군집이 서로 겹치지 않게 군집의 수를 결정하기 때문에 분할적 군집 분석이라고 함. 계층적 군집 방법에 비해 계산량이 적기 때문에 대용량 데이터 분석에 적합한 방법 (예 : k-means clustering)
* 계층적 군집 분석 : 한 군집 내 부분 군집을 허용하는 방법으로 각각의 관측치를 하나의 군집으로 간주하고 데이터의 특성이 가까운 부분 군집끼리 순차적으로 결합해 나가는 병합적 방법 (Agglomerative method) 이 주로 사용됨. 덴드로그램을 이용한 시각화를 통해 각 관측치가 어떤 관측치와 가까운지 파악할 수 있어 군집들을 자연적인 계층 구조로 분석하고자 할 때 유용함

### 군집 분석 : 주요 이슈
* 군집 분석에 활용할 변수들의 척도 차이를 확인. 군집 분석은 거리를 기반으로 군집을 형성하기 때문에 활용 변수의 척도 차이가 큰 경우 다ㄴ위가 큰 변수에 가중치가 부여되어 군집이 왜곡될 수 있음 -> 각 변수의 척도 차이가 존재하면 분석에 앞서 정규화 또는 표준화 작업을 수행해야 함
* 계층적 군집은 어떤 유사성 측도와 연결 방법으로 이용할 것인지 결정하는 것이 군집 형성에 큰 영향을 줄 수 있음. 또한 군집 결정을 위한 어디에서 덴드로 그램의 절단 계층을 선택하는 것도 중요한 의사결정 요인
* 분할적 군집 형성 방법 중 하나인 k=평균 군집 분석은 k 값을 어떻게 지정해줄 것인가에 대한 선택이 군집 형성에 매우 큰 영향을 줌. k- 평균 군집 분석 역시 각 관측치 간 유사성 측도를 무엇으로 할지 결정하는 것 또한 중요함
*  -> 사전 결정 사항을 결정하기 위한 보편 타당한 방법은 따로 존재하지 않음! 데이터의 성향이나 산업 특성에 맞게 타당한 방법 선택이 중요하며 경우에 따라선 다양한 시도를 통해 해답을 찾아야 함

### 군집 분석 : 고려 사항
* 군집 분석은 데이터 변동에 민감함 -> 관측치를 하나만 제거해도 군집이 달라질 수 있음
* k-평균 군집 방법이나 계층적 군집 방법 모두 모든 관측치에 군집을 할당함
* 데이터 수집 오류 또는 이상 현상으로 발생한 데이터 (outlier)가 포함된 경우 어디에도 속하지 않지만 알고리즘에 의해 불가피하게 군집에 속하게 되는 관측치가 발생할 수 있음
* 이런 경우 해당 관측치로 인해 군집이 왜곡될 수 있음
-> 모둔 관측치에 군집을 할당하는 것이 적절하지 않을 수 있음 이런 경우 혼합 모델(EM, expectation-maximization -> GMM, Gausian Mixture Modeling)을 이용해 문제를 개선할 수 있음

### 군집 분석 : 수치형 거리 측도
* Norm : 벡터의 크기 or 길이 측정 함수
* 민코우스키 거리 : Norm의 일반화 표현
* 유클리드 거리 : 두 개체 i와 j사이의 거리(2-norm)
* 맨하탄 거리 : 1-norm
* 상관 계수 활용 : Pearson, Cosine (Eisen), Kendall
* 마할라노비스 : 변수 분산과 공분산 구조

### K-평균 군집화 : 최적 K
* 전역 최적이 아닌 국소 최적을 찾기 떄문에 초기 군집에 따라 결과가 달라짐
* 군집 초기값을 다르게 하여 여러 번 실행한 뒤 각 결과를 비교해 목적 함수를 최소화
* 사전에 산점도 등을 통해 데이터 특성을 확인해 데이터가 뭉쳐 있는 그룹의 수를 k값으로 선택하는 방법 활용(변수 2개!)
* 변수의 숫자가 많은 경우 산점도 행렬을 이용해 각 변수들의 관계를 통해 특성을 파악하거나 주성분 분석이나 오토 - 인코더(auto-encoder)와 같은 차원 축소 방법을 이용해 변수의 수를 줄여 특성이나 방법을 파악하는 방법을 이용할 수 있음

### K - medoids 군집화
* K-means는 중심으로부터 거리를 기반으로 군집화하기 때문에 구형으로 뭉쳐져 있는 볼록한 데이터 셋에는 비교적 잘 적용되나, 오목한 형태의 군집 모델의 특성을 구별해 내는데 문제를 보임
* 동 떨어져 있는 데이터나 노이즈에 매우 민감하게 반응하며, 사전에 클러스터 개수를 정해야 하는 것도 단점. 초기 클러스터를 어떻게 선택하느냐에 따라 글로벌 최소값에 도달하지 못하는 경우도 있음
* 클러스터의 중심을 좌표평면상 임의의 점이 아니라 데이터셋의 값 중 하나를 선정해 실제 데이터셋에 있는 값을 중심점으로 하기 때문에 이상치와 노이즈 처리가 우수하고, 매우 강건(robust)하게 수렴하는 방법

### 계층적 군집화 : 개요
* 한 군집 내에 부분 군집을 혀용하는 방법으로 덴드로그램을 이용한 시각화를 활용해 군집의 수를 결정할 수 있다는 장점이 잇음
* 계층적 군집은 부분 군집을 허용하기 위해 전체 데이터를 하나의 군집으로 하고 이를 부분 군집으로 나눈 후, 각 부분 군집을 다시 나누는 계층 형식의 군집화 방법을 이용
* 사전에 군집 수를 정할 필요가 없음
* 덴드로그램을 이용해 시각적으로 이해가 쉽고 적당한 군집 수 선정이 용이
* 계층적 군집은 위사결정나무의 생성 과정과 역순으로 계층 구조를 만듦. 개별 관측치에서 시작해 이들을 묶어 나가는 방법을 이용하며 이를 하향식 응집이라고 함

### 계층적 군집화 : 분할적 군집화와의 비교
* 계층적 군집분석은 군집 특성상 분할적 군집분석보다 많은 정보를 파악할 수 있음
* 각 관측치들이 얼마나 가까운지 이질적인지 등에 대한 정보를 시각적으로 보다 상세하게 알 수 있어 유용
* But 계층적 군집분석이 분할적 군집분석보다 좋은 것은 아님  --> 계층적 군집 분석은 특성들이 뒤석여 부정확한 결과 얻을 위험 있음

### 계층적 군집화 : 연결 종류
* 단일 연결법 : 군집의 관측치들의 거리를 계산, 그리고 계산된 거리 중 가장 작은 값을 두 군집의 거리로 결정. 각 군집의 거리가 가장 작은, 즉 유사도가 높은 군집을 다음 단걔에서 하나의 군집으로 묶음
* 완전 연결법 : 단일 연결법과 달리 각 군집 내 관측치 거리 간 제일 큰 값을 ㄱㄱ 군집의 거리로 취급
* 평균 연결법 : 각 군집 내 관측치들이 평균 거리를 군집의 거리로 취급
* 중심 연결법 : 각 군집 내 모든 관측치 간의 거리를 계산하고 이 거리의 무게 중심을 군집들 간의 거리로 취급. 각 군집이 덴드로그램에서 개별 클러스터 중 어느 하나보다 낮은 높이에서 융합될 수 있음 (Inversion 문제)
* 와드 방법  각 군집이 합쳐졌을 때 중심 군집(군집 내 관측치들의 평균)과 각 관측치 간의 차이 제곱을 거리로 이용. 만약 주어진 두 군집이 서로 이질적이라면, 이 두 군집을 병합할 경우 군집 중심과 각 관측치들의 거리가 상당히 커지기 때문에 유용하게 이용할 수 있는 방법

### 계층적 군집화 : 결과의 평가 척도
* 응집도 (Cohesion): 하나의 군집 안에 있는 레코드들이 얼마나 근접해 있는가 나타내는 지표
* 격리도 (Seperation) : 하나의 군집이 다른 군집들과 얼마나 분리되어 있는지를 나타내는 지표
* 실루엣 계수 (Silhouette coefficient) : 해당 데이터가 얼마나 같은 군집 내의 데이터와 얼마나 가깝게 군집화 돼 있고, 다른 군집에 있넌 데이터와는 얼마나 멀리 분리돼 있는지를 나타내는 지표
* ---> 군집 내 응접도를 최소화, 격리도를 최대화, 실루엣 계수가 1에 가까울수록 우수한 군집
